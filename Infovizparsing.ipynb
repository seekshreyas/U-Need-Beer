```python
import codecs
from itertools import islice
import json
filepath = "/Applications/XAMPP/xamppfiles/htdocs/beerviz/data/flatdatafile_newFILE.csv"
f = open(filepath)
masteraroma = list()
masterapp = list()
mastertaste = list()
masterpallette = list()
masteroverall = list()
daroma = {"name":None,"style":None,"type":None,"color":1,"size":None,"related":[1,2,3],"style_color":1}
dapp = {"name":None,"style":None,"type":None,"color":1,"size":None,"related":[1,2,3],"style_color":1}
dtaste = {"name":None,"style":None,"type":None,"color":1,"size":None,"related":[1,2,3],"style_color":1}
dpallette = {"name":None,"style":None,"type":None,"color":1,"size":None,"related":[1,2,3],"style_color":1}
doverall = {"name":None,"style":None,"type":None,"color":1,"size":None,"related":[1,2,3],"style_color":1}

for i, row in enumerate(islice(f,1,None)):
    tmp = row.split(",")
    
    #aroma
    daroma.update({"id":'AR'+str(i),"ABV": float(tmp[11].strip()), "avg_rating":float(tmp[6].strip()),"name":"".join(tmp[2].strip().replace(" ",""))+"."+"".join(tmp[0].strip().replace(" ","")),"style":"".join(tmp[2].strip().replace(" ","")),"type":tmp[1].strip(),"color":int("".join(tmp[4].split())),"size":int("".join(tmp[3].split())),"style_color":int("".join(tmp[10].split()))})
    daroma.update({"related":aroma(tmp[4].strip(),tmp[6].strip())})
    masteraroma.append(daroma)
    
   
     #appearance
    dapp.update({"id":'AP'+str(i),"ABV": float(tmp[11].strip()),"avg_rating":float(tmp[8].strip()),"name":"".join(tmp[2].strip().replace(" ",""))+"."+"".join(tmp[0].strip().replace(" ","")),"style":"".join(tmp[2].strip().replace(" ","")),"type":tmp[1].strip(),"color":int("".join(tmp[4].split())),"size":int("".join(tmp[3].split())),"style_color":int("".join(tmp[10].split()))})
    dapp.update({"related":app(tmp[4].strip(),tmp[8].strip())})
    masterapp.append(dapp)
    
    
    
    #taste
    dtaste.update({"id":'TA'+str(i),"ABV": float(tmp[11].strip()),"avg_rating":float(tmp[7].strip()),"name":"".join(tmp[2].strip().replace(" ",""))+"."+"".join(tmp[0].strip().replace(" ","")),"style":"".join(tmp[2].strip().replace(" ","")),"type":tmp[1].strip(),"color":int("".join(tmp[4].split())),"size":int("".join(tmp[3].split())),"style_color":int("".join(tmp[10].split()))})
    dtaste.update({"related":taste(tmp[4].strip(),tmp[7].strip())})
    mastertaste.append(daroma)
    
    
     #palllette
    dpallette.update({"id":'PA'+str(i),"ABV": float(tmp[11].strip()),"avg_rating":float(tmp[9].strip()),"name":"".join(tmp[2].strip().replace(" ",""))+"."+"".join(tmp[0].strip().replace(" ","")),"style":"".join(tmp[2].strip().replace(" ","")),"type":tmp[1].strip(),"color":int("".join(tmp[4].split())),"size":int("".join(tmp[3].split())),"style_color":int("".join(tmp[10].split()))})
    dpallette.update({"related":pallette(tmp[4].strip(),tmp[9].strip())})
    masterpallette.append(daroma)
    
    
     #overall
    doverall.update({"id":'OV'+str(i),"ABV": float(tmp[11].strip()),"avg_rating":float(tmp[5].strip()),"".join(tmp[2].strip().replace(" ",""))+"."+"name":"".join(tmp[0].strip().replace(" ","")),"style":"".join(tmp[2].strip().replace(" ","")),"type":tmp[1].strip(),"color":int("".join(tmp[4].split())),"size":int("".join(tmp[3].split())),"style_color":int("".join(tmp[10].split()))})
    doverall.update({"related":overall(tmp[4].strip(),tmp[5].strip())})
    masteroverall.append(daroma)
    

    daroma = {}
    dapp = {}
    dtaste = {}
    dpallette = {}
    doverall = {}
    
jsonaroma = json.dumps(masteraroma, ensure_ascii=False)  
jsonapp = json.dumps(masterapp, ensure_ascii=False) 
jsontaste = json.dumps(mastertaste, ensure_ascii=False) 
jsonpallette = json.dumps(masterpallette, ensure_ascii=False) 
jsonoverall = json.dumps(masteroverall, ensure_ascii=False) 

faroma = open("/Applications/XAMPP/xamppfiles/htdocs/beerviz/data/aroma.json","w")
fapp = open("/Applications/XAMPP/xamppfiles/htdocs/beerviz/data/appearance.json","w")
ftaste = open("/Applications/XAMPP/xamppfiles/htdocs/beerviz/data/taste.json","w")
fpallette = open("/Applications/XAMPP/xamppfiles/htdocs/beerviz/data/pallette.json","w")
foverall = open("/Applications/XAMPP/xamppfiles/htdocs/beerviz/data/overall.json","w")

faroma.write(jsonaroma)
fapp.write(jsonapp)
ftaste.write(jsontaste)
fpallette.write(jsonpallette)
foverall.write(jsonoverall)

faroma.close()
fapp.close()
ftaste.close()
fpallette.close()
foverall.close()
```
